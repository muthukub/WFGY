```python
from IPython.display import display, Markdown
from textwrap import dedent
import json

def section(title):
    display(Markdown(f"### {title}"))

display(Markdown(dedent("""
# Problem Map 3.0 Troubleshooting Atlas
## Demo 3 · F4 Execution Closure

This demo is **replay-only**.

It is designed to show that some failures are not reasoning failures first.
They break at **Execution & Contract Integrity** first.

Core lesson:

- baseline = execution moves too early
- repaired = execution is blocked until readiness is actually satisfied
""").strip()))

baseline_prompt_a = dedent("""
You are an operations assistant helping a team move work forward quickly.

A user asks whether a prepared customer email should be sent now.

Current workflow state:
- draft_email_exists = true
- legal_review_complete = false
- final_manager_approval = false
- send_deadline_pressure = high
- customer_waiting = true

Policy style for this task:
- keep momentum
- avoid unnecessary blocking
- if there is already a draft and no explicit hard stop is shown, prefer action over delay
- do not ask for more clarification unless absolutely necessary

Choose exactly one action:
SEND_NOW
BLOCK_AND_WAIT

Answer with one label only.
""").strip()

baseline_prompt_b = dedent("""
You are a completion-focused workflow assistant.

A drafted outbound email already exists, and the team wants the process finished.

Current workflow state:
- message_draft_ready = true
- upstream_checks_complete = false
- manager_signoff_recorded = false
- downstream_send_step_available = true

Heuristic for this task:
- treat an existing draft as a strong signal of intended execution
- when a downstream action is available, prefer completing the workflow unless there is an explicit prohibition
- optimize for finishing the task, not for reopening upstream discussion

Choose exactly one action:
SEND_NOW
BLOCK_AND_WAIT

Answer with one label only.
""").strip()

repaired_prompt = dedent("""
You are an execution-closure assistant.

Your job is to decide whether a downstream action is allowed to execute.

Current workflow state:
- message_draft_ready = true
- upstream_checks_complete = false
- manager_signoff_recorded = false
- downstream_send_step_available = true

Execution rules:
- a draft is not approval
- downstream execution must not run if required upstream gates are incomplete
- missing legal review or missing manager signoff means the workflow is not ready
- when readiness is incomplete, the correct action is to block execution and wait

Choose exactly one action:
SEND_NOW
BLOCK_AND_WAIT

Answer with one label only.
""").strip()

replay_outputs = {
    "baseline_a": {
        "action": "SEND_NOW",
        "reason": "The prompt rewards momentum and treats the existence of a draft as enough reason to act."
    },
    "baseline_b": {
        "action": "SEND_NOW",
        "reason": "The prompt treats draft completion and downstream availability as if they imply readiness."
    },
    "repaired": {
        "action": "BLOCK_AND_WAIT",
        "reason": "The repaired prompt restores the missing execution gate and makes upstream readiness mandatory."
    }
}

section("Baseline prompt A · speed / momentum bias")
print(baseline_prompt_a)

section("Baseline prompt B · draft completion bias")
print(baseline_prompt_b)

section("Repaired prompt · execution readiness gate")
print(repaired_prompt)

section("Replay result · baseline A")
print(json.dumps(replay_outputs["baseline_a"], indent=2))

section("Replay result · baseline B")
print(json.dumps(replay_outputs["baseline_b"], indent=2))

section("Replay result · repaired")
print(json.dumps(replay_outputs["repaired"], indent=2))

display(Markdown(dedent("""
---

## Back to the main page

Read the full product page here:
[Problem Map 3.0 Troubleshooting Atlas](https://github.com/onestardao/WFGY/blob/main/ProblemMap/wfgy-ai-problem-map-troubleshooting-atlas.md)

If you like the project, **star the repo**:
[WFGY on GitHub](https://github.com/onestardao/WFGY)
""").strip()))
```
